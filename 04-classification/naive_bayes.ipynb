```python
def normalize(wine):
    result = wine.copy()
    for feature_name in wine.columns:
        max_value = wine[feature_name].max()
        min_value = wine[feature_name].min()
        result[feature_name] = (wine[feature_name] - min_value) / (max_value - min_value)
    return result
    
red=normalize(df)

red.head()
```

```python
aggs = df.groupby('quality').agg(['mean', 'std'])
aggs
```

```python

# P(A|C) - for single example (with distinct feature from distinct group)
def cond_prob(row, feature, group):
    mu = aggs[feature]['mean'][group] # mean
    std = aggs[feature]['std'][group] # standard deviation

    # one selected observation (sigle value)
    sel_obs = df.iloc[row][feature]

    prob = stats.norm.pdf(sel_obs, loc=mu, scale=std)

    return prob
    

```

```python
def predict_class(row):
    classes = []

    min_class = np.amin(df['quality'].values)
    max_class = np.amax(df['quality'].values)

    for group in range(min_class, max_class):
        # P(C)
        prob = len(df[df['quality'] == group])/len(df.index)
        for feature in df.columns:
            if feature == 'quality': break
            # P(C) * P(A|C)
            prob *= cond_prob(row, feature, group)
        classes.append(prob)
    return np.argmax(classes) + min_class # from what number classes starts
    
```
